```python
import pandas as pd
import numpy as np
import datetime
```

Load dataset and prepare it for analysis

```python
csv_file = r"netflix_titles.csv"
df = pd.read_csv(csv_file)
df.shape # Get dataset shape
```

```python
# Drop rows with no entries in columns
df = df.dropna()
df.shape # Get dataset shape after removing bad data
df.date_added = pd.to_datetime(df.date_added) # convert add date to datetime format
# filter by years
# df = df.loc[(df.date_added >= '2008-01-01') & (df.date_added < '2018-01-01')] # < 2018
# df = df.loc[(df.date_added >= '2018-01-01')] # >= 2018
```

```python
# One-hot encode 'country' and 'listed_in' columns
countries = df['country'].str.get_dummies(sep=',').add_prefix('Country_')
genres = df['listed_in'].str.get_dummies(sep=',').add_prefix('Genre_')
# Edit column names and combine duplicates
countries.columns = countries.columns.str.replace('_ ', "_")
countries = countries.groupby(countries.columns, axis=1).sum()
genres.columns = genres.columns.str.replace('_ ', "_")
genres = genres.groupby(genres.columns, axis=1).sum()
```

```python
# Concatenate One-hot encoded dummy frames to main dataframe
df = pd.concat([df, countries], axis=1).drop('country', axis=1)
df = pd.concat([df, genres], axis=1).drop('listed_in', axis=1)
df.shape # New shape after concatenation
```

## Data Analysis Phase

```python
# Count number of movies & shows for each country
country_sum = countries.sum(axis=0)
# plot 20 most popular movies
country_sum = country_sum.sort_values(ascending=False)
country_sum = country_sum[0:20]
country_sum.plot(kind='bar', title="Top 20 countries in the Netflix Dataset")
# country_sum.to_csv("country_sum.csv") # export country frequencies for presentation
# might check the country sums for different years
```

```python
# Count number of movies & shows for each genre
genre_sum = genres.sum(axis=0)
# plot 10 most popular genres
genre_sum = genre_sum.sort_values(ascending=False)
genre_sum = genre_sum[0:10]
genre_sum.plot(kind='bar', title="Top 10 genres in the Netflix Dataset")
# genre_sum.to_csv("genre_sum.csv") # export genre frequencies for presentation
```

```python
# after running to check % change between movies before 2018 and after
# the country with the most growth is india with a growth rate of 430%
df_india = df.loc[df.Country_India == 1] # filter rows
genres_india = df_india[genres.columns.to_list()] # extract genre columns
# Count number of movies & shows for each genre
genre_sum_india = genres_india.sum(axis=0)
# plot 10 most popular genres
genre_sum_india = genre_sum_india.sort_values(ascending=False)
genre_sum_india = genre_sum_india[0:10]
genre_sum_india.plot(kind='bar', title="Top 10 genres in india")
# genre_sum_india.to_csv("genre_sum_india.csv") # export genre frequencies for presentation
# might check the country sums for different years/periods
```
